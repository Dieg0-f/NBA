# Estrazione 200 fotogrammi per l'annotazione per creare il modello

```python
import cv2
import os
import random
```

```python
# Percorso della cartella con i video
cartella_video = "/home/diego/Documents/GitHub/NBA/Video_bos"
# Cartella di output per i fotogrammi
cartella_output = "/home/diego/Documents/GitHub/NBA/Computer_vision/annotation_frames"
os.makedirs(cartella_output, exist_ok=True)

# Estensioni video supportate
estensioni_video = ['.mp4', '.avi', '.mov', '.mkv']

# Trova tutti i file video nella cartella
tutti_i_video = [
    os.path.join(cartella_video, f) for f in os.listdir(cartella_video)
    if os.path.isfile(os.path.join(cartella_video, f)) and any(f.lower().endswith(est) for est in estensioni_video)
]

# Prendi 200 video casuali (o meno, se ce ne sono di meno)
video_selezionati = random.sample(tutti_i_video, min(200, len(tutti_i_video)))

# Cicla sui video selezionati
for percorso_file in video_selezionati:
    nome_file = os.path.basename(percorso_file)

    # Carica il video
    cap = cv2.VideoCapture(percorso_file)
    
    success, fotogramma = cap.read()
    if success:
        # Salva il primo fotogramma
        nome_output = os.path.splitext(nome_file)[0] + "_frame1.jpg"
        percorso_output = os.path.join(cartella_output, nome_output)
        cv2.imwrite(percorso_output, fotogramma)
        print(f"Salvato: {percorso_output}")
    
    cap.release()
```
