```python
import json
import numpy as np
import pandas as pd
```

```python
with open('Scraping_data/shot_chart_bos.json', 'r') as f:
    shot_chart = json.load(f)

df_shot = pd.DataFrame(shot_chart)
df_shot['action_type'] = 'Shot'
df_shot
```

```python
with open('Scraping_data/play_by_play_bos.json', 'r') as f:
    play_by_play = json.load(f)

df_play = []
for game_data in play_by_play:
    df_play.extend(game_data)

# Crea il dataframe pandas dalla lista appiattita
df_play = pd.DataFrame(df_play)
df_play
```

```python
df_merged_multi_col = pd.merge(df_play, df_shot,
                               left_on=['gameID', 'quarter', 'time', 'is_home_team', 'action_type'],
                               right_on=['gameID', 'quarter', 'timer', 'is_home_team', 'action_type'],
                               how='left')

df_merged_multi_col = df_merged_multi_col.drop(columns='timer')

pd.set_option('display.max_columns', None)
df_merged_multi_col
```

```python
import pandas as pd

# Funzione per determinare se mantenere la riga o eliminarla
def filter_rows(row):
    home_team = False if row['gameID'][:3] == 'bos' else True
    is_home_team = row['is_home_team']
    return (home_team + is_home_team) in [0, 2]

# Applicare il filtro e mantenere solo le righe che soddisfano la condizione
df = df_merged_multi_col[df_merged_multi_col.apply(filter_rows, axis=1)]
df
```

```python
df.loc[:, 'videoID'] = df.apply(
    lambda row: f"{row['gameID']}_{row['quarter']}_{row['time']}.mp4" if row['action_type'] == 'Shot' else None, axis=1
)
df
```
