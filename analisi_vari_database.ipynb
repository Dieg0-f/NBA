***Analisi Video scaricati e elaborati***

```python
import os
import pandas as pd
import json

# Percorso alla cartella dei video scaricati
video_folder = '/home/diego/Documents/GitHub/NBA/Video_bos'
downloaded_videos = set(os.listdir(video_folder))

# Caricamento del DataFrame
# df_bos = pd.read_csv(...)  # Se non è già caricato

# Aggiunta colonna 'download': 1 se scaricato, 0 se no
df_bos['download'] = df_bos['videoID'].apply(lambda x: 1 if x in downloaded_videos else 0)

# Caricamento file JSON con i risultati dell'elaborazione
with open('Computer_vision/shot_frame_extraction_results.json', 'r') as f:
    extraction_data = json.load(f)['risultati_per_video']

# Aggiunta colonna 'extracted'
def check_extraction(video_id):
    if video_id in extraction_data:
        return 1 if extraction_data[video_id] > 0 else 0
    else:
        return None  # Non presente nei risultati

df_bos['extracted'] = df_bos['videoID'].apply(check_extraction)
```
